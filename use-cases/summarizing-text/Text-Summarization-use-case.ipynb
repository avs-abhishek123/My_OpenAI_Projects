```python
# creating the df
summarization_df = pd.DataFrame(columns=[
    'Sl_no', 
    'Model_name', 
    'Input', 
    'Inp_word_count' , 
    'Output', 
    'Out_word_count', 
    'Accuracy', 
    'Remarks'
    ])
```

<center>
<h4 style="color: blue; font-size: 26px;">
    <b>
            Add a row of data to dataframe 
    </b>
</h4>
</center>

```python
def add_row_to_df(
    df,
    Sl_no = 0 , 
    Model_name = "Sample Model Name" , 
    Input = "Sample Input Text", 
    Inp_word_count = 0, 
    Output = "NA", 
    Out_word_count = 0, 
    Accuracy = 0 , 
    Remarks = "Sample Row"):
  new_rows = {
        'Sl_no' : Sl_no,
        'Model_name' : Model_name,
        'Input' : Input,
        'Inp_word_count' : Inp_word_count,
        'Output' : Output,
        'Out_word_count' : Out_word_count,
        'Accuracy' : Accuracy,
        'Remarks' : Remarks
  }
  updated_df = df.append(new_rows, ignore_index = True)
  return updated_df
```
